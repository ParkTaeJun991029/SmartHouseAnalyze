# 전체 데이터 가져오기

```python
def init_path():
    global device_path
    global light_module_path
    global temperature_module_path
    global illuminance_module_path
    global presence_module_path
    global device_list
    global light_module_list
    global temperature_module_list
    global illuminance_module_list
    global presence_module_list
device_path='D:/주거_조명/smartgateway/device/'
light_module_path='D:/주거_조명/smartgateway/light/'
temperature_module_path='D:/주거_조명/smartgateway/temperature/'
illuminance_module_path='D:/주거_조명/smartgateway/illuminance/'
presence_module_path='D:/주거_조명/smartgateway/presence/'
device_list=os.listdir(device_path)
light_module_list=os.listdir(light_module_path)
temperature_module_list=os.listdir(temperature_module_path)
illuminance_module_list=os.listdir(illuminance_module_path)
presence_module_list=os.listdir(presence_module_path)

def init_df():
    global device_df
    global light_module_df
    global temperature_module_df
    global illuminance_module_df
    global presence_module_df

device_df= pd.DataFrame()
light_module_df= pd.DataFrame()
temperature_module_df= pd.DataFrame()
illuminance_module_df= pd.DataFrame()
presence_module_df= pd.DataFrame()

#실행시키기
#디바이스
for n in tqdm(device_list):
    device_df_tmp=pd.read_csv(device_path+n)
    device_df=pd.concat([device_df,device_df_tmp])
    device_df=device_df.drop('Unnamed: 0',axis=1)
    device_df=device_df.reset_index(drop=True)

        
#light_module
for n in tqdm(light_module_list):
    light_df_tmp=pd.read_csv(light_module_path+n)
    light_module_df=pd.concat([light_module_df,light_df_tmp])
    light_module_df=light_module_df.drop('Unnamed: 0',axis=1)
    light_module_df=light_module_df.reset_index(drop=True)


#temperature_module
for n in tqdm(temperature_module_list):
    temperature_df_tmp=pd.read_csv(temperature_module_path+n)
    temperature_module_df=pd.concat([temperature_module_df,temperature_df_tmp])
    temperature_module_df=temperature_module_df.drop('Unnamed: 0',axis=1)
    temperature_module_df=temperature_module_df.reset_index(drop=True)



#illuminance_module
for n in tqdm(illuminance_module_list):
    illuminance_df_tmp=pd.read_csv(illuminance_module_path+n)
    illuminance_module_df=pd.concat([illuminance_module_df,illuminance_df_tmp])
    illuminance_module_df=illuminance_module_df.drop('Unnamed: 0',axis=1)
    illuminance_module_df=illuminance_module_df.reset_index(drop=True)


#presence_module
for n in tqdm(presence_module_list):
    presence_df_tmp=pd.read_csv(presence_module_path+n)
    presence_module_df=pd.concat([presence_module_df,presence_df_tmp])
    presence_module_df=presence_module_df.drop('Unnamed: 0',axis=1)
    presence_module_df=presence_module_df.reset_index(drop=True)

#gid
light_gid=list(set(light_module_df['gid']))
temperature_gid=list(set(temperature_module_df['gid']))
illuminance_gid=list(set(illuminance_module_df['gid']))
presence_gid=list(set(presence_module_df['gid']))
device_gid=list(set(device_df['gid']))

#did
light_did=list(set(light_module_df['did']))
temperature_did=list(set(temperature_module_df['did']))
illuminance_did=list(set(illuminance_module_df['did']))
presence_did=list(set(presence_module_df['did']))
device_did=list(set(device_df['did']))
```
